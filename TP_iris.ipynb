```python
# import load_iris function from sklearn.datasets library
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import statistics

iris = load_iris()
X = iris.data
y = iris.target
print(iris['feature_names'])
```

```python
variance_empirique = []
a = 0
for classe in range(max(y)+1):
    for desc in range(len(X[0])):
        a += statistics.variance(X[y==0,desc])    
    variance_empirique.append(round(a / len(X[0]),3))
    
print(variance_empirique)    
```

```python
moyennes_globale = []
variances_globale = []
for desc in range(len(X[0])):
    print(iris['feature_names'][desc], end=':')
    print(" M:", round(statistics.mean(X[:,desc]), 3) , end = ' |')
    print(" V:", round(statistics.variance(X[:,desc]), 3))
    moyennes_globale.append(round(statistics.mean(X[:,desc]),3))
    variances_globale.append(round(statistics.variance(X[:,desc]),3))
print("\nMoyennes globales: ", moyennes_globale)
print("Variances globales: ", variances_globale)
```

```python
moyennes_inter = []
v_inter = []
indice = 0

#'a' repr√©sente les moyennes intra, qui vont nous permettre d'obtenir la moyenne inter
a = 0
b = 0
for desc in range(len(X[0])):
    for classe in range(max(y)+1):                           
        a += statistics.mean(X[y==classe,desc])
        b += statistics.variance(X[y==classe,desc])                
    moyennes_inter.append(round(a / len(X[0]),3))
    v_inter.append(round(b / len(X[0]),3))
    a = 0
    b = 0

print("moyennes interclasses: \n\t", moyennes_inter)  
print("variances interclasses: \n\t", v_inter)
```

```python
moyennes_intra = []
v_intra = []
indice = 0

for classe in range(max(y)+1):
    print("\nclasse ", iris['target_names'][classe])
    for desc in range(len(X[0])):    
        print(iris['feature_names'][desc], end= ':')             
        moyennes_intra.append(round(statistics.mean(X[y==classe,desc]), 3))
        v_intra.append(round(statistics.variance(X[y==classe,desc]),3))
        
        print(" M:", moyennes_intra[indice], end='|' )
        print(" V:", v_intra[indice])
        indice += 1
        
moyennes_intra = np.array(moyennes_intra)
moyennes_intra = moyennes_intra.reshape(max(y)+1, len(X[0])) 

print("\nMoyennes Intra : \n", moyennes_intra)

v_intra = np.array(v_intra)
v_intra = v_intra.reshape(max(y)+1, len(X[0]))

print("\nVariances intra : \n", v_intra)
```

```python
class MyKnn:
    def __init__(self, _k):
        self.k = _k
        self.train = None
        self.labels = None
        
    def apprentissage(self, X, y):
        '''
        lazy algorithm
        '''
        self.train = X
        self.labels = y
        
    def predict(self, X):
        predictions = []
        for test_sample in X:  # iteration sur chaque exemple test                    
            distances = []
            for train_sample in self.train:  # itearation sur chaque exemple train                                             
                distances.append(distance_euclidienne(test_sample, train_sample))
                
            nearest_neighbors = np.argsort(distances)          
            k_nearest_classes = []
            
            for i in range(self.k):
                k_nearest_classes.append(self.labels[nearest_neighbors[i]])
        
            predictions.append(vote_majoritaire(k_nearest_classes))
        return np.array(predictions)
```

```python
def distance_euclidienne(pt1, pt2):
    if len(pt1) != len(pt2):
        raise ValueError('inputs must have the same dimensions')
    return np.sqrt(np.sum((pt1 - pt2)**2))
```

```python
def vote_majoritaire(classes):
    labels, counts = np.unique(classes, return_counts=True) 
    return labels[np.argmax(counts)]
```
